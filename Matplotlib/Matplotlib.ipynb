# Pie Graph

```python
x = np.arange(4)
car1 = [40, 17, 28, 12]
car2 = [34, 21, 24, 15]
```

```python
x
```

```python
plt.bar(x, car1, width=0.3, tick_label=[2017,2018,2019,2020], label='Car 1')
plt.bar(x+0.3,car2, width=0.3, label='Car 2')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Sales")
plt.title("Sales graph")
plt.xlim(-1,4)
plt.ylim(0,50)
```

# PIE Graph

```python
subjects = ["Physics", "Chemistry", "Maths", "English"]
marks = np.array([69, 79, 90, 95])
```

```python
plt.pie(x = marks, labels = subjects, explode=[0,0.2,0,0], shadow = True, autopct="%0.2f")
```

# Subplots

```python
x = np.arange(0,10, 0.1)
y1 = x**2
y2 = x**3
y3 = np.sin(x)
y4 = np.cos(x)
```

```python
plt.subplot(2,2,1)
plt.plot(x,y1)

plt.subplot(2,2,2)
plt.plot(x,y2)

plt.subplot(2,2,3)
plt.plot(x,y3)

plt.subplot(2,2,4)
plt.plot(x,y4)

plt.show()
```

# 3D Plots

```python
a = np.arange(4)
b = np.arange(4,7)
```

```python
a
```

```python
b
```

```python
a,b = np.meshgrid(a,b)
```

```python
a
```

```python
b
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection ='3d')
ax.plot_surface(a,b,a+b)
plt.show()
```

```python
a = np.arange(-1,1,0.005)
b = a
a,b = np.meshgrid(a,b)

fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.plot_surface(a,b, a**2 + b**2)
```
