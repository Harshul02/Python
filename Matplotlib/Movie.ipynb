```python
movie = pd.read_csv('./movie_meta.csv')
```

```python
def preprocess(ele):
    return len(ele.strip())
```

```python
movie['title_length'] = movie['movie_title'].apply(preprocess)
```

```python
movie.columns
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.hist(movie['title_length'], bins = 50)
plt.xticks(range(0,100,5))
plt.xlabel('Movie Title Length')
plt.ylabel('Frequency of Movies')
plt.show()
```

```python
freq_title = {}

for l in movie['title_length']:
    if freq_title.get(l) is None:
        freq_title[l] = 1
    else:
        freq_title[l] +=1
```
