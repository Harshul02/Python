## Merging DataFrames

```python
df1 = pd.DataFrame({'S_Name': ['Harshul', 'Aayush', 'Shravan', 'Amit'], 'CGPA' : [3,4,5,6]})
```

```python
df1
```

```python
df2 = pd.DataFrame({'T_Name': ['Harshul', 'Aayush', 'Shravan', 'Amit'], 'CGPA': [3,6,8,9]})
```

```python
df2
```

```python
df1.merge(df2, how='inner')
```

```python
df1.merge(df2, how='left')
```

```python
df1.merge(df2, how='outer')
```
