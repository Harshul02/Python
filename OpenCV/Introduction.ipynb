### Importing Library
<mark>import cv2</mark>

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Reading and Displaying Images

```python
img = cv2.imread("test1.jpg")
```

```python
type(img)
```

```python
np.shape(img)
```

```python
plt.imshow(img)
```

```python
img
```

```python
cv2.imshow("People", img)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```

### Changing ColorSpaces

```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
img_rgb.shape
```

```python
plt.imshow(img_rgb)
```

### Image Resizing

```python
img_rgb.shape
```

```python
plt.imshow(img_rgb)
```

```python
img_new = cv2.resize(img_rgb, (500,500))
```

```python
img_new.shape
```

```python
plt.imshow(img_new)
```

```python
cv2.imshow("People", img_new)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

### Blur

```python
img_blur = cv2.blur(img_rgb, (50,50))
```

```python
plt.imshow(img_blur)
```

### Image Rotation

```python
img_rgb.shape
```

```python
rows, cols = img_rgb.shape[:-1]
```

```python
M = cv2.getRotationMatrix2D((rows/2, cols/2), 10, 1)
```

```python
i = cv2.warpAffine(img_rgb, M, (cols,rows))
```

```python
plt.imshow(i)
```

### Image Translation

```python
M = np.float32([[1,0,50], [0,1,20]])
```

```python
M
```

```python
i = cv2.warpAffine(img_rgb, M, (cols, rows))
```

```python
plt.imshow(i)
```

### Edge Detection
        - Canny Edge Detection

```python
i = cv2.Canny(img_rgb, 50, 50)
```

```python
plt.imshow(i)
```
