# Face Detection

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = plt.imread("test1.jpg")
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
detector = cv2.CascadeClassifier('./haarcascade_frontalface_default.xml')
```

```python
all_faces = detector.detectMultiScale(img, 1.1, 5)
```

```python
all_faces.shape
```

```python
x,y,w,h = all_faces[0]
```

```python
img = cv2.rectangle(img,(x,y), (x+w, y+h), (0,0,0), 5)
```

```python
plt.imshow(img)
```

```python
for face in all_faces:
    x,y,w,h = face
    img = cv2.rectangle(img, (x,y), (x+w, y+h), (255,0,0), 5)
```

```python
plt.imshow(img)
```
